```python
from transformers import pipeline
import pandas as pd
```

```python
pipe = pipeline("text-classification", model="finiteautomata/bertweet-base-sentiment-analysis")
```

```python
df = pd.read_csv("/content/data.csv")
```

```python
from random import randint
```

```python
results = {'dataset_sentiment': [], 'pipeline_sentiment': [], 'score': []}
n = 1
goal = 100
while n<=goal:
    rand = randint(0, 5841)
    piperes = pipe(df['Sentence'][rand])
    print(df['Sentence'][rand])
    results['dataset_sentiment'].append((df['Sentiment'][rand])[0:3].upper())
    results['pipeline_sentiment'].append(piperes[0]['label'])
    # TODO work on getting the score
    results['score'].append(piperes[0]['score'])
    n+=1
```
